```python
import pandas as pd
import numpy as np

df=np.array(pd.read_csv("results.csv"))

prob=df[:,1:8]

class_dist=np.zeros(prob.shape[1])
pred=np.argmax(prob, axis=1)

for i in range(prob.shape[1]):
    class_dist[i]=len(np.where(pred==i)[0])
    
class_dist=np.array([int(class_dist[i]) for i in range(class_dist.shape[0])])

ths=0.9

a_ths=np.where(prob>=0.9)
ind_ths=a_ths[0]
class_ths=a_ths[1]

class_dist_ths=np.zeros(prob.shape[1])

for i in range(prob.shape[1]):
    class_dist_ths[i]=len(np.where(class_ths==i)[0])
```

```python
one=np.where(class_ths==0)[0]
two=np.where(class_ths==1)[0]
four=np.where(class_ths==2)[0]
```

```python
df1=pd.DataFrame(df[ind_ths[one]])
df2=pd.DataFrame(df[ind_ths[two]])
df4=pd.DataFrame(df[ind_ths[four]])
```

```python
import glob

l=glob.glob("/data/kburdge/ZTF/Lightcurves/*")
```

```python
mean_mag1=np.zeros(df1[0].shape[0])
for i in range(df1[0].shape[0]):
    tempdf=pd.read_csv(l[int(df1[0][i])], header=None, delimiter=" ")
    mean_mag1[i]=np.round(np.mean(tempdf[1]),4)
    
mean_mag2=np.zeros(df2[0].shape[0])
for i in range(df2[0].shape[0]):
    tempdf=pd.read_csv(l[int(df2[0][i])], header=None, delimiter=" ")
    mean_mag2[i]=np.round(np.mean(tempdf[1]),4)
    
mean_mag4=np.zeros(df4[0].shape[0])
for i in range(df4[0].shape[0]):
    tempdf=pd.read_csv(l[int(df4[0][i])], header=None, delimiter=" ")
    mean_mag4[i]=np.round(np.mean(tempdf[1]),4)
```

```python
df1_ra_dec=pd.DataFrame(np.transpose([df1[8],df1[9]]))
df2_ra_dec=pd.DataFrame(np.transpose([df2[8],df2[9]]))
df4_ra_dec=pd.DataFrame(np.transpose([df4[8],df4[9]]))

df1_ra_dec[2]=mean_mag1
df2_ra_dec[2]=mean_mag2
df4_ra_dec[2]=mean_mag4
```

```python
df1.to_csv("EW_prob0.9.csv", header=['og_ind','EW','EA','RRab','RRc','RRd','RSCVn','LPV','RA','Dec'])
df2.to_csv("EA_prob0.9.csv", header=['og_ind','EW','EA','RRab','RRc','RRd','RSCVn','LPV','RA','Dec'])
df4.to_csv("RRab_prob0.9.csv", header=['og_ind','EW','EA','RRab','RRc','RRd','RSCVn','LPV','RA','Dec'])
```

```python
df1_ra_dec.to_csv("EW_prob0.9.csv", header=['RA','Dec', 'Mean_Mag'])
df2_ra_dec.to_csv("EA_prob0.9.csv", header=['RA','Dec', 'Mean_Mag'])
df4_ra_dec.to_csv("RRab_prob0.9.csv", header=['RA','Dec', 'Mean_Mag'])
```
