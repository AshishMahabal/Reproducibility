```python
import numpy as np

radec=np.load("lcid.npy")

ra_dec=np.array([radec[i][30:None] for i in range(radec.shape[0])])

ra=np.array([ra_dec[i].split("_")[0] for i in range(ra_dec.shape[0])])

dec=np.array([ra_dec[i].split("_")[1] for i in range(ra_dec.shape[0])])

np.save("RA",ra)
np.save("Dec",dec)
```

```python
import pandas as pd

df=pd.read_csv("test.csv")

df['RA']=ra
df['Dec']=dec

df.to_csv("results.csv")
```
